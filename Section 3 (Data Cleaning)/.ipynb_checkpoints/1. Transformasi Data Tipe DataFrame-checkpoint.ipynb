## 4. Group By

```python
group1 = diabetes[['Age','BloodPressure']].groupby(['Age']).count()
group1.head()
```

```python
group2 = diabetes[['Age','BloodPressure']].groupby(['Age']).mean()
group2.head()
```

## Join

```python
tblclients = pd.read_excel('data/products.xlsx',sheet_name='tblclients')
tblproducts = pd.read_excel('data/products.xlsx',sheet_name='tblproducts')
```

```python
tblclients
```

```python
tblproducts
```

```python
pd.merge(tblproducts,tblclients,left_on='userid',right_on='id',how='inner')
```

```python
tblproducts.merge(tblclients,left_on='userid',right_on='id',how='inner')
```

## Union

```python
## union by rows
r1 = diabetes.iloc[:4,:]
r2 = diabetes.iloc[4:10,:]
```

```python
r1.head()
```

```python
r2.head()
```

```python
union1 = pd.concat([r1,r2],axis=0,ignore_index=True)
union1.head(10)
```

```python
## union by columns
c1 = diabetes[['Pregnancies','Glucose','BloodPressure']]
c2 = diabetes[['SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']]
```

```python
c1.head()
```

```python
c2.head()
```

```python
union2 = pd.concat([c1,c2],axis=1,ignore_index=True)
union2.columns = c1.columns.values.tolist() + c2.columns.values.tolist()
union2.head()
```

## Not IN, IN

```python
tblproducts.head()
```

```python
ls_prod = ['rice','milk']
```

```python
tblproducts[tblproducts['product'].isin(ls_prod)]
```

```python
tblproducts[~tblproducts['product'].isin(ls_prod)]
```

## Apply Function

```python
diabetes.head()
```

```python
def RangeUmur(age):
    cat = []
    if (age > 20 & age < 30):
        cat = 'Middle'
    elif age >= 30:
        cat = 'Old'
    else:
        cat = 'Nan'
    return cat
```

```python
diabetes['category_umur'] = diabetes['Age'].apply(RangeUmur)
```
