## Apply Function

```python
diabetes.head()
```

```python
def RangeUmur(age):
    cat = []
    if (age > 20 & age < 30):
        cat = 'Middle'
    elif age >= 30:
        cat = 'Old'
    else:
        cat = 'Nan'
    return cat
```

```python
diabetes['Age'].apply(RangeUmur)
```

```python
# Membuat variabel baru bernama cluster dan sekaligus assign value nya
diabetes['Cluster'] = diabetes['Age'].apply(RangeUmur)
diabetes
```

```python
cat_umur = diabetes['Age'].apply(RangeUmur)
```

```python
diabetes['category_umur'] = cat_umur
```
